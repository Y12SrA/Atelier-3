```python
import requests
from bs4 import BeautifulSoup

def get_page_content(url):
    page = requests.get(url)
    if page.status_code == 200:
        return page.content
    else:
        print(f"Failed to retrieve page content. Status code: {page.status_code}")
        return None

def get_text_score(text):
    return len(text) / 100  # Adjust the scoring logic as needed

def main():
    url = "https://www.islamweb.net/ar/fatwa/3682/%D9%81%D8%B1%D8%A7%D8%A6%D8%B6-%D8%A7%D9%84%D9%88%D8%B6%D9%88%D8%A1-%D8%A7%D9%84%D9%85%D8%AA%D9%81%D9%82-%D8%B9%D9%84%D9%8A%D9%87%D8%A7-%D9%88%D8%A7%D9%84%D9%85%D8%AE%D8%AA%D9%84%D9%81-%D9%81%D9%8A%D9%87%D8%A7"
    src = get_page_content(url)

    if src:
        soup = BeautifulSoup(src, "lxml")
        details = []
        ship = soup.find_all("div", {'class': 'right-nav'})

        def get_info(ship):
            title = ship.contents[0].find('h2').text.strip()
            all_info = ship.contents[0].find_all('p')
            text = " ".join([info.text.strip() for info in all_info])
            score = get_text_score(text)
            details.append({"Text": text, "Score": score})

        get_info(ship[0])

        # Print or save the details
        for idx, detail in enumerate(details, start=1):
            print(f"Text {idx} (Arabic Language):")
            print(detail["Text"])
            print(f"Score: {detail['Score']:.2f}")
            print()

if __name__ == "__main__":
    main()
```
